```python
for col in chennai_cars.columns[:4]:
    chennai_cars[col] = chennai_cars[col].apply(ast.literal_eval)

chennai_cars
```

```python
car_detail_df = pd.json_normalize(chennai_cars['new_car_detail'])
```

```python
car_detail_df
```

```python
car_overview_df = pd.json_normalize(chennai_cars['new_car_overview'])
```

```python
car_overview_df
```

```python
flattened_data = []
for index, row in car_overview_df.iterrows():
    columns_dict = {item['key']: item['value'] for item in row['top']}
    flattened_data.append(columns_dict)

car_overview_flattened = pd.DataFrame(flattened_data)
car_overview_flattened
```

```python
car_feature_df = pd.json_normalize(chennai_cars['new_car_feature'])
car_feature_df
```

```python
flattened_data2 = []

for index, row in car_feature_df.iterrows():
    # Extract the 'value' from each dictionary in the 'top' column list
    features = [item['value'] for item in row['top']]
    
    # Join the features into a single string (or keep as a list if needed)
    feature_str = ', '.join(features)
    
    # Append the feature string to the flattened data list
    flattened_data2.append(feature_str)

car_top_feature_flattened = pd.DataFrame(flattened_data2, columns=['Top_Feature'])
car_top_feature_flattened
```

```python
flattened_data3 = []

# Iterate through the nested data
for index, row in car_feature_df.iterrows():
    row_data = {}  # To hold the flattened data for each row
    
    # Iterate through each category in the row
    for category in row['data']:
        heading = category['heading']
        values = [item['value'] for item in category['list']]
        values_str = ', '.join(values)
        
        # Store the values string under the heading key
        row_data[heading] = values_str
    
    # Append the row's dictionary to the flattened data list
    flattened_data3.append(row_data)

car_feature_indetail_flattened = pd.DataFrame(flattened_data3)
car_feature_indetail_flattened
```

```python
car_specs_df = pd.json_normalize(chennai_cars['new_car_specs'])
car_specs_df
```

```python
flattened_data4 = []
for index, row in car_specs_df.iterrows():
    columns_dict = {item['key']: item['value'] for item in row['top']}
    flattened_data4.append(columns_dict)

car_top_specs_flattened = pd.DataFrame(flattened_data4)
car_top_specs_flattened
```

```python
flattened_data5 = []
for index, row in car_specs_df.iterrows():
    row_data = {}
    for category in row['data']:
        for item in category['list']:
            key = item['key']  # Extract the key (e.g., 'Engine', 'Max Power', etc.)
            value = item['value']  # Extract the corresponding value (e.g., '998 CC', '118.36bhp')
            row_data[key] = value
    flattened_data5.append(row_data)

flattened_df = pd.DataFrame(flattened_data5)

car_specs_flattened = pd.concat([car_top_specs_flattened, flattened_df], axis=1)
car_specs_flattened
```

```python
chennai_cars_df = pd.concat([car_detail_df, car_overview_flattened, car_top_feature_flattened, car_feature_indetail_flattened, car_specs_flattened], axis = 1)
chennai_cars_df.head()
```

```python
chennai_cars_df['City'] = 'Chennai'
chennai_cars_df.head()
```

```python
chennai_cars_df.to_csv('chennai_cars_df.csv')
```
